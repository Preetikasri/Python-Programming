###  1. Rock, Paper, Scissors

```python
q=["Rock", "Paper", "Scissors"]
import random
def rps(c,h):
	""" Returns the game result """
	if c==h:
		return"Draw"
	elif c=="Rock":	
		if h=="Paper":
			return'Human Win'
		elif h=="Scissors":
			return"Computer Win"
	elif c=="Paper":
		if h=="Rock":
			return'Computer Win'
		elif h=="Scissors":
			return'Human Win'
	elif c=="Scissors":
		if h=="Rock":
			return'Human Win'
		elif h=="Paper":
			return"Computer Win"
a=0
b=0
c=0
i=0
win=0
d={}
h_choice=None
while h_choice!="Quit":
	l= sorted(d.items()) # List of key,value pairs sorted by keys
	if i==0 or l[0][1]==l[1][1]==l[2][1]: # Random choice for first game
		c_choice=random.choice(q)
	else:                                       # Computer brains
		if l[0][1]>l[1][1] and l[0][1]>l[2][1]:
			c_choice="Scissors"
		elif l[1][1]>l[0][1] and l[1][1]>l[2][1]:
			c_choice="Paper"
		elif l[0][1]==l[1][1] and l[0][1]>l[2][1]:
			c_choice="Paper"
		elif l[0][1]==l[2][1] and l[0][1]>l[1][1]:
			c_choice="Scissors"
		elif l[1][1]==l[2][1] and l[1][1]>l[0][1]:
			c_choice="Rock"
		else:
			c_choice="Rock"
	h_choice=raw_input("Choose Rock, Paper or Scissors to play. Type Quit to stop playing:") # Evaluating user input
	if h_choice!="Rock" and h_choice!="Paper" and h_choice!="Scissors" and h_choice!="Quit":
		print"Not a valid input"
		continue
	elif h_choice=="Quit":
		print"Games Played:",i
		print"Games won:",win
	r=rps(c_choice,h_choice)
	i+=1 # Iteration for games played
	if r=="Draw":
		print"Draw",":",c_choice,"draws with",h_choice
	elif r=="Human Win":
		print"Human Win:",h_choice,"beats",c_choice
		win+=1 # Iteration for games won by human
	elif r=="Computer Win":
		print"Computer Win:",c_choice,"beats",h_choice
	if h_choice=='Rock': # Updating dictionary
		a+=1
	elif h_choice=="Paper":
		b+=1
	elif h_choice=="Scissors":
		c+=1
	d['Rock']=a
	d['Paper']=b
	d['Scissors']=c


	
			
        
			
```

### 2. Voters in Florida

```python
import re

fp = open('../data/FloridaVoters.html', 'r')
text = fp.read()

text = text.replace(',','') #get rid of commas - we need this to create integers

matches = re.findall('<td>(.*?)</td>',text) #parse the html text to only include the table

#create lists of the first, second, and third rows of table
x = matches[::6]
y = matches[1::6]
z = matches[2::6]

#make the numbers into integers instead of strings
z = map(int, z)
y = map(int, y)

#zip the lists into tuples of three
voters = zip(x,y,z)

#sort by dem voters
sorted(voters, key=lambda x: x[2])
```

## 3. Google of Quotes

### (a) 

```python
fp  = open('../data/quotes.txt', 'r')
fp.seek(0)  ## To start the pointer at the begining of the file pointer. 
mylist = []
#reading the data
l = fp.readline()
l = l.rstrip("\n")
while l is not '':
    mylist.append(l)
    l = fp.readline()
    l = l.strip("\n")
# Combining Author/Speaker and their respective quotes
i = 0
fin_list=[]
while (i <= (len(mylist)-4)):
    #print i
    s = mylist[i] + "-" + mylist[i+1]
    fin_list.append(s)
    i = i+2
print(fin_list)
```

## (b)

```python
import re 

def words_in_full_quotes(quotes):
    
    words = re.split('\W*',quotes.lower())
    return words

### Demonstrative example
words_in_full_quotes('If Jack`s in love, he`s no judge of Jill`s beauty.-Benjamin Franklin')
```

## (c)

```python
def get_d2(quote):
    q = words_in_full_quotes(quote)
    d = {}
    for k in q:
        d[k] = q.count(k)
    return d

def posit_dictionary(corpus):
	pos_dict = {}
	for k in corpus:
	    d2 = get_d2(k)
	    pos_dict[k] = d2
	return pos_dict


## Demonstrative Example

postings_list_dictionary=posit_dictionary(fin_list)
postings_list_dictionary
```

## (d)

```python
reverse_posting_list = {}

for i in range(len(fin_list)):
    
    quote = fin_list[i]
    
    for word in words_in_full_quotes(quote):
        
        if word not in reverse_posting_list.keys():
            reverse_posting_list[word]={}
            reverse_posting_list[word][quote] = 1
                
        else:
            if quote not in reverse_posting_list[word]:
                reverse_posting_list[word][quote] = 1
            else:
                reverse_posting_list[word][quote] += 1

print("Reverse Posting List dictionary : \n")
reverse_posting_list
```

## (e)

```python
def tf(word, quote, dict1):
    
    if (quote in dict1):
        if(word in dict1[quote]):
            numr = dict1[quote][word]
            #print(numr)
            denom = max(dict1[quote].values())
            #print(denom)
            TF = (numr)/(denom*1.0)
            return TF
        else:
            return(" Word not found in Quote!!")
    else:
        return ("Quote not found in Corpus!!")

## Defining function to get IDF

import math
def IDF(word, dict1, dict2):
    if word in dict2:
        numr = len(dict1.keys())
        #print(numr)
        denom = len(dict2[word].keys())
        #print(denom)
        IDF = math.log(numr/denom)
        return IDF
    else:
        return ("word not found !! ")

def TF_IDF(word, quote, dict1, dict2):
    tf_ = tf(word, quote, dict1)
    idf = IDF(word, dict1, dict2)
    if((isinstance(tf_, float)) & (isinstance(idf, float))):
        tfidf = tf_*(idf*1.0)
        return tfidf
    else:
        return ("Errr...Either word or quote or their combination not present in the corpus!!")
    


## Demonstrative Example

TF_IDF('change','Observe constantly that all things take place by change, and accustom thyself to consider that the nature of the Universe love nothing so much as to change. The Universe is change.-Marcus Aurelius',
      postings_list_dictionary, reverse_posting_list)
```

## (f)

```python
def search_single_word(word):
    fin_dict = {}
    dict_key = reverse_posting_list[word].keys()
    for key in dict_key:
        fin_dict[key] =  TF_IDF(word, key, postings_list_dictionary, reverse_posting_list)
    return fin_dict

## Demonstrative Example

search_single_word('morning')
```

## (g)

```python
def Quote_search_multiple_words(inp):
    res_d = {}
    for w in inp:
        #print("word here is: ",w)
        ky = reverse_posting_list[w].keys()
        #res_d = {}
        for j in ky:
            if(j not in res_d.keys()):
                res_d[j] = TF_IDF(w, j, postings_list_dictionary, reverse_posting_list)
            else: 
                #print(j)
                #print("first TFIDF is:",res_d[j])
                res_d[j] = res_d[j] + TF_IDF(w, j, postings_list_dictionary, reverse_posting_list)

    return res_d

## Demonstrative Example
Quote_search_multiple_words(['men', 'love', 'life'])
```
